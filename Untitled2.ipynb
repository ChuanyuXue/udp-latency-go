```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def read_data(path):
    data = pd.read_csv(path,header=None)
    data.columns = ['index','length','t1','t2','t3','t4']
    data = data[data['index'] > 0]
    data = data.sort_values('index')
    data['rtt'] = data['t4'] - data['t1'] - (data['t3'] - data['t2'])
    data['time'] = (data['t1'] - np.uint64(data[data.index == 1]['t1'])) / 1e9
    return data

data.append(read_data("tight.csv"))
print(np.mean(data[-1].rtt))

data = []
for i in range(1, 9):
    data.append(read_data("%d000more.csv"%i))
    print(np.mean(data[-1].rtt))
```

```python
plt.figure(figsize=(20,15))
for i, df in enumerate(data):
    sns.lineplot(x = 'time', y = 'rtt', label = 'Add %de3 ns more'%(i), data = df)
```

```python
plt.figure(figsize=(20,15))
for kk, df in enumerate(data):
    series = list(df.rtt)
    result = []
    for i, v in enumerate(series):
        if i > 25 and i < len(series) - 25:
            result.append(np.average(series[i-25:i+25]))
    sns.lineplot(x = range(len(result)), y = result, label = 'Add %de3 ns more'%(kk),)
```
