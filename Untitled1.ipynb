```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def read_data(path):
    data = pd.read_csv(path,header=None)
    data.columns = ['index','length','t1','t2','t3','t4']
    data = data[data['index'] > 0]
    data = data.sort_values('index')
    data['rtt'] = data['t4'] - data['t1'] - (data['t3'] - data['t2'])
    data['time'] = (data['t1'] - np.uint64(data[data.index == 1]['t1'])) / 1e9
    return data

data = []
for i in range(1, 10):
    data.append(read_data("exp2_tight/%de5early.csv"%i))
    print(np.median(data[-1].rtt))
```

```python
plt.figure(figsize=(20,15))
for i, df in enumerate(data):
    sns.lineplot(x = 'time', y = 'rtt', label = 'Defer %de5 ns'%(i+1), data = df)
```
